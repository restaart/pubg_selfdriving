```python
import sys

if '..' not in sys.path:
    sys.path.append('..')

from modules import direct_input

import time
import threading
import keyboard
```

```python
class KeyThread(threading.Thread):
    def __init__(self, key,period=0.1,duty=0):
        super().__init__()
        self.keyCode = direct_input.keyCodes[key]
        self.period = period
        self.duty = duty
        self.enabled = True
        
    @property
    def duty(self):
        return self.__duty
    
    @duty.setter
    def duty(self,value):
        if value < 0:
            value = 0
        elif value > 1:
            value = 1
        self.__duty = value
        if value == 0:
            direct_input.ReleaseKey(self.keyCode)
        elif value == 1:
            direct_input.PressKey(self.keyCode)  
       
    def run(self):
        while(self.enabled):
            if(0<self.duty<1):
                direct_input.PressKey(self.keyCode)
                time.sleep(self.period*self.duty)
                direct_input.ReleaseKey(self.keyCode)
                time.sleep(self.period*(1-self.duty))        
            
            
            
```

```python
class KeyWatchThread(threading.Thread):
    def __init__(self, key,period=0.001):
        super().__init__()
        self.key = key
        self.recordedData = []
        self.enabled = True
        self.period = period
        
    def run(self):
        zero_time = time.time()
        while(self.enabled):
            self.recordedData.append((keyboard.is_pressed(self.key),time.time()-zero_time))
            time.sleep(self.period)
        
```

```python
kt = KeyThread('w',period=0.1,duty=0)
kwt = KeyWatchThread('w',0.001)
time.sleep(1)
kt.start()
kwt.start()

kt.duty = 0.8
time.sleep(1)

kt.duty = 0.2
time.sleep(1)

kt.duty = 1
time.sleep(1)

kt.duty = 0.0
time.sleep(1)


kt.enabled = False
kwt.enabled = False
kt.join()
kwt.join()
```
