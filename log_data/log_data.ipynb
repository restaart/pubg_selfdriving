```python
import time
import numpy as np
import matplotlib.pyplot as plt
import keyboard
from PIL import ImageGrab

from modules.minimap_gps import minimap_gps as GPS
from modules import speed_detector,azimuth_detector

import pandas as pd
```

```python
gps = GPS(map_name='camp_jackal')
```

```python
def getScreenShot(bbox=(0,0,1920,1080)):
    return np.array(ImageGrab.grab(bbox))



keyboard.wait('F7')
print('started')
ss = []
xx = []
yy = []
aa = []
tt = []
while not keyboard.is_pressed('F8'):
    img = getScreenShot()
    speed = speed_detector.speed_from_image(img)
    x,y = gps.get_position(img)
    azimuth = azimuth_detector.azimuth_from_image(img)
    
    ss.append(speed)
    xx.append(x)
    yy.append(y)
    aa.append(azimuth)
    tt.append(time.time())
    

    
    
```

```python
plt.plot(tt,aa)
plt.show()
```

```python
df = pd.DataFrame()
df['speed'] = ss
df['azimuth'] = aa
df['x'] = xx
df['y']  = yy
df['time'] = tt

df.to_csv('dacia_big_log.csv',index=None)
```

```python
df  = df.fillna(method ='ffill')
```

```python
s = df.speed*1000/3600
a = df.azimuth
x = [df.x[0]]
y = [df.y[0]]
t = df.time
for i in range(1,len(s)):
    dt = t[i] - t[i-1]
    dx = s[i] * np.sin(a[i]*np.pi/180)*dt
    dy = s[i] * np.cos(a[i]*np.pi/180)*dt
    x.append(x[-1] + dx)
    y.append(y[-1] + dy)
    
x = np.array(x)
y = np.array(y)
```

```python
plt.plot(x,y-1200)
plt.plot(df.x,-df.y)
plt.show()
```
