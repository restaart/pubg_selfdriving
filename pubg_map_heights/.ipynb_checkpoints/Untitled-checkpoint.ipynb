```python
class JSONDataObject(object):
    
    def __init__(self,data_):
        self.values = data_
    
    def __dir__(self):
        if(type(self.values) is list):
            if (type(self.values[0]) is dict):
                keys = set()
                for val in self.values:
                    if val is not None:
                        keys.update(val.keys())         
                return list(keys)
            else:
                return []#JSONDataObject(self.values[0]).__dir__()
        elif(type(self.values) is dict):
            return list(self.values.keys())
        else:
            return []
        
    def __exrtact_values(self,d,name):
        return [None if di is None else (di[name] if name in di else None) for di in d]
    
    def __getattr__(self, attr):        
        if type(self.values) is list :
            result = JSONDataObject(self.__exrtact_values(self.values,attr))
        else:
            result = JSONDataObject(self.values[attr])
            
        if(len(result.__dir__())>0):
            return result
        else:
            return result.values
    
```
