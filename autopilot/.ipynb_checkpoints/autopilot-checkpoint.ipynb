```python
def draw_points(points):
    points = np.array(points)
    map_points = (points/gps.map_meters_per_pixel)
    plt.figure(figsize=(7,7))
    plt.imshow(gps.map_image[100:250,:200])
    plt.scatter(map_points[:,0],map_points[:,1]-100,s=3,c=[1,2],cmap='tab10')
    plt.show()
    
```

```python
next_point_offset_meters = 20
curpoint = position

def update_waypoint(curpoint, path,next_point_offset_meters):
    next_point_ix = np.argmin(np.abs(np.abs(np.abs(path - curpoint).sum(axis=1) - next_point_offset_meters)))
    next_point = path[next_point_ix]
    path =  path[next_point_ix:]
    return next_point,path

next_point,path = update_waypoint(curpoint,path,next_point_offset_meters)
next_point
```

```python

def get_azimuth_to_point(curpoint,next_point):
    return -np.arctan2(*(next_point-curpoint))/np.pi*180 + 180

next_point,path = update_waypoint(curpoint,path,next_point_offset_meters)
draw_points([next_point,curpoint])

curpoint = curpoint + 0.3*(next_point-curpoint)

azimuth_to_point = get_azimuth_to_point(curpoint,next_point)
print(azimuth_to_point)
```

```python
a1 = 357
a2 = 175
def angle_substruct(a1, a2):

    a1,a2 = np.array([a1,a2])/180*np.pi  
    a1 = a1 % (2*np.pi)
    a2 = a2 % (2*np.pi)

    if((a1 - a2) > np.pi):
        
        return (2*np.pi - a1 + a2)/np.pi*180

    if((a2 - a1)> np.pi):

        return -(2*np.pi - a2 + a1)/np.pi*180  

    return (a2 - a1)/np.pi*180  

angle_substruct(a1,a2)
```

```python
class PID():
    def __init__(self,p,i,d):
        self.p = p
        self.i = i
        self.d = d
        self.lastError = None
        self.intError = 0
        
    def control(self,error,dt):
        if self.lastError is not None:
            derror = (self.lastError - error)/dt
        else:
            derror = 0
        self.lastError = error
        self.intError += error*dt
        
        val = error*self.p + derror*self.d + self.intError*self.i
               
        return val
```

```python
class double_key_control(object):
    def __init__(self,key1,key2,pid1,pid2):
        self.right_control = KeyThread(key1,period=0.1,duty=0)
        self.left_control  = KeyThread(key2,period=0.1,duty=0)
        
#         self.right_control.start()
#         self.left_control.start()
        
        self.right_pid = pid1
        self.left_pid  = pid2
        
    def control(self,error,dt):
        if(error >= 0):
            control_value = self.right_pid.control(error,dt)
            self.left_control.duty = 0
            self.right_control.duty = control_value
        else:
            control_value = self.left_pid.control(-error,dt)
            self.left_control.duty = control_value
            self.right_control.duty = 0
            
        print(self.right_control.duty,self.left_control.duty)
            
#     def __del__(self): 
#         self.left_control.join()
#         self.right_control.join()
```

```python
kw_a = KeyWatchThread('w')
kw_d =  KeyWatchThread('s')

t = 0
dt = 0.1

val = 300
targ = 45
time.sleep(2)
control = double_key_control('w','s',PID(0.01, 0.1, 0),PID(0.01, 0.1, 0))
kw_a.start()
kw_d.start()
while t < 10:
    t+=dt
    err = angle_substruct(targ,val)
    val+=err*0.1
    val = val%360
    control.control(err,dt)
    
del double_key_control    
kw_a.join()
kw_d.join()
```
